# **Pregunta 3:**

Se tienen las siguientes listas de datos: (**ejecuta** la celda a continuación)

```
import pandas as pd
list1 = {'Lima': 8751700,'La Libertad': 1837000,'Piura': 1829500,'Cajamarca': 1525100,'Puno': 1402500,'Junín': 1341100,
         'Cusco': 1308800,'Arequipa': 1273200,'Lambayeque':1250300,'Ancash':1142400,'Loreto':1029000,'Callao':1000000}
list2 = {'Lima': 34802,'La Libertad': 25500,'Piura': 35893,'Cajamarca': 33318,'Puno': 71999,'Junín': 44327,
         'Cusco': 71987,'Arequipa': 63345,'Lambayeque':14231,'Ancash':35915,'Loreto':368852,'Callao':147}
tabla = pd.DataFrame({'poblacion': list1,'area': list2})
tabla
```

Realiza un **gráfico** donde puedas identificar cuáles son los departamentos con **mayor y menor población**. Repite el mismo gráfico para comparar las **áreas**.

```
# Se puede identificar los departamentos con mayor y menor población utilizando un gráfico de barras
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.bar(tabla.poblacion.index,tabla.poblacion.values)
plt.xlabel('Departamentos')
plt.ylabel('Poblacion')
plt.title('Poblacion de Principales Departamentos')
# Imprimir valores de Población en el centro de cada barra
for i in range(len(tabla.poblacion.values)):
        plt.annotate(str(tabla.poblacion.values[i]), xy=(tabla.poblacion.index[i],tabla.poblacion.values[i]), ha='center', va='bottom')

# Lima es el departamento con mayor población
# Callao es el departamento con menor población
```

```
# Se puede identificar los departamentos con mayor y menor población utilizando un gráfico de barras
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.bar(tabla.area.index,tabla.area.values)
plt.xlabel('Departamentos')
plt.ylabel('Area')
plt.title('Area de Principales Departamentos')
# Imprimir valores de Población en el centro de cada barra
for i in range(len(tabla.area.values)):
        plt.annotate(str(tabla.area.values[i]), xy=(tabla.area.index[i],tabla.area.values[i]), ha='center', va='bottom')

# Loreto es el departamento con mayor area
# Callao es el departamento con menor area
```

# **Pregunta 4**

Genera una **tabla** donde indiques la población y área **promedio**.

```
df = tabla
df['average_area'] = tabla['area'].mean()
df['density'] = round(df['poblacion']/df['area'],3)
df = df.drop(['area'],axis=1)
df
```

# **Pregunta 5**

Finalmente puedes concluir si el tamaño de la población o el área **influye** para considerar a un departamento **importante**. Redacta tu respuesta.

```
# Se puede identificar los departamentos con mayor y menor densidad poblacion utilizando un gráfico de barras
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.bar(df.density.index,df.density.values)
plt.xlabel('Departamentos')
plt.ylabel('Densidad Poblacional')
plt.title('Densidad Poblacional de Principales Departamentos')
# Imprimir valores de Población en el centro de cada barra
for i in range(len(df.density.values)):
        plt.annotate(str(df.density.values[i]), xy=(df.density.index[i],df.density.values[i]), ha='center', va='bottom')
```

Considero que en este caso, los campos de **población** y **área** si influyen porque nos permiten conocer sobre la densidad poblacional de cada departamento.
Podemos concluir lo siguiente:


*   **Callao presenta un alto nivel de densidad poblacional**: Indica que es un departamento con población concentrada
*   **Loreto presentan un bajo nivel de densidad poblacional**: Indica que es un departamento con población dispersa
